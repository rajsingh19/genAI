Function Calling Explanation

Function calling is just like calling a waiter in a restaurant.
When we go to a restaurant, we call the waiter and give our order (input), such as the food item we want. The waiter takes our order, goes to the kitchen, and brings back the food (output).

In the same way, in programming, we call a function and provide it with some input (called arguments). The function performs the required task using that input and then gives us the result (output).

Just like we receive food only after placing the order, we get the required output only after calling the function.

#### Example

```python
def check_even_odd(num):
    if num % 2 == 0:
        return "Even"
    else:
        return "Odd"

# Example
number = int(input("Enter a number: "))
result = check_even_odd(number)
print("The number is:", result)
```

#### Handling Error

```python
def check_even_odd(num):
    if num < 0:
        return "Error: Please enter a positive number."
    elif num % 2 == 0:
        return "Even"
    else:
        return "Odd"

number = int(input("Enter a number: "))
print(check_even_odd(number))
```

## Recursive Function

By a Recursive Function we can say that a Function is
going to call itself again and again

```python
## Factorial of a Number
## factorial of 8 ---> 8*7*6*5*4*3*2*1
## In factorial a base case is the condition that stops the recursion.
```

### Factorial of a num

```python
def factorial(n):
    # Base case
    if n == 0 or n == 1:
        return 1
    
    # Recursive case
    return n * factorial(n - 1)


# Example
num = 5
print("Factorial of", num, "is", factorial(num))
```

#### FUNCTION WITH MULTIPLE PARAMETER

```python
def add_numbers(a, b, c):
    return a + b + c

print(add_numbers(2, 3, 4))
```

#### *args (or *num) ‚Üí allows passing unlimited positional arguments to a function.

```python
# function to add multiple numbers
def addition_positional(*num):
    add = 0
    for i in num:
        add += i
    print("The addition of the numbers is:", add)

# Calling the function
addition_positional(10, 20, 30)
addition_positional(5, 15)
```
