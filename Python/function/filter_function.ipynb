# Filter Function

filter() is used to select specific elements from a list based on a condition.

It keeps only the elements where the condition is True.

### Syntax

filter(function, iterable)

function → condition (must return True/False)

iterable → list, tuple, etc.

#### Example 1: Filter Even Numbers

```python
numbers = [1, 2, 3, 4, 5, 6]

result = filter(lambda x: x % 2 == 0, numbers)
print(list(result))
```

#### Example 2: Filter Numbers Greater Than 10

```python
numbers = [5, 12, 7, 18, 3]

result = filter(lambda x: x > 10, numbers)
print(list(result))
```

#### Example 3: Filter Strings Starting with "A"

```python
names = ["Aman", "Rahul", "Anita", "Priya"]

result = filter(lambda name: name.startswith("A"), names)
print(list(result))
```

#### Example 4: Using Normal Function

```python
def is_positive(x):
    return x > 0

numbers = [-2, -1, 0, 1, 2]

result = filter(is_positive, numbers)
print(list(result))
```
