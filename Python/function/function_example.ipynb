 #### 1.To create a function that converts temperature from celcius to Fahrenheit or Fahrenheit to Celcius

````python
def conv_fahrenheit_to_celsius(num):
    return (num - 32) * 5/9
result = float(input("number:"))  
print(conv_fahrenheit_to_celsius(result))  
````

````python
def conv_celsius_to_fahrenheit(num):
    return (num * 9/5) + 32
result = float(input("number:"))  
print(conv_celsius_to_fahrenheit(result))  
````

#### single function for both 

````python
def convert_temperature(value, unit):
    if unit == "C":
        return (value * 9/5) + 32   # Convert to Fahrenheit
    elif unit == "F":
        return (value - 32) * 5/9   # Convert to Celsius
    else:
        return "Invalid unit"

print(convert_temperature(20, "C"))  # 68.0
print(convert_temperature(68, "F"))  # 20.0
````

## define a Function that would behave as a Password Strength Checker
```
Strong password-->
if your password:
1.length is greater than 8 characters
2.password must consist of atleast lowercase character (a,b,c... z)
3.password must consist of atleast a uppercase character (A,B,C... Z)
4.password must consists of at least a digit (1, 2,)
5.password must consists of at least a special character(!@#$%^&)

Weak passwords-->
If a password doesnt match any above condition then it will be considered as a weak password
```

````python
def check_password(password):
    
    # Condition 1: Length greater than 8
    if len(password) <= 8:
        return "Weak Password"
    
    # Condition 2: At least one lowercase letter
    has_lower = any(char.islower() for char in password)
    
    # Condition 3: At least one uppercase letter
    has_upper = any(char.isupper() for char in password)
    
    # Condition 4: At least one digit
    has_digit = any(char.isdigit() for char in password)
    
    # Condition 5: At least one special character
    special_characters = "!@#$%^&"
    has_special = any(char in special_characters for char in password)
    
    # Check all conditions
    if has_lower and has_upper and has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"
````

````python
print(check_password("Abc@12345"))   # Strong Password
print(check_password("abc123"))      # Weak Password
````

###  Check that whether if the current string is a palindrome or not

````python
def is_palindrome(text):
    # Convert to lowercase (optional, for case-insensitive check)
    text = text.lower()
    # Remove spaces
    text = text.replace(" ", "")
    # Check if string is same as its reverse
    if text == text[::-1]:
        return "Palindrome"
    else:
        return "Not Palindrome"

````

````python
print(is_palindrome("Madam"))
````

````python
print(is_palindrome('A man a plan a canal panama'))
````

### Calculate the Total Cost of items in a shopping cart

````python
# Calculate the Total Cost of items in a shopping cart

cart = [
    {'item': 'Apple', 'Cost': 5, 'Quantity': 20},
    {'item': 'Banana', 'Cost': 4, 'Quantity': 15},
    {'item': 'Guava', 'Cost': 3, 'Quantity': 10}
]

# Function to calculate total cost
def cart_total(cart_items):
    Total_cost = 0
    
    for item in cart_items:
        Total_cost += item['Cost'] * item['Quantity']
    
    print("The Total cost for the Cart items is:", Total_cost)


# Calling the function
cart_total(cart)
````
