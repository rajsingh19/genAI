# Lamda Function

A lambda function is a small, anonymous (no name) function written in one line.

#### Syntax

```python
lambda arguments : expression
```

Arguments --> Input values for our Function

Expression ---> Logic ( Function Body )

No def

No return

Only one expression

### ðŸ”¹ Normal Function:

```python
def add(a, b):
    return a + b
```

### ðŸ”¹ Lambda Version:

####  Example 1: Addition

```python
add = lambda a, b: a + b
```

#### Example 2: Square of a Number

```python
square = lambda x: x * x
print(square(4))
```

#### Example 3: Check Even or Odd

```python
even = lambda x: "Even" if x % 2 == 0 else "Odd"
print(even(10))
```
