## write the content into our file line by line --> writelines()

```python
lines = [
    "Hello\n",
    "Welcome to Python\n",
    "File Handling Example\n"
]

with open("sample.txt", "w") as file:
    file.writelines(lines)
```

Write the content into the binary file --> wb

Read the content from the binary file  --> rb


| Mode   | Meaning      |
| ------ | ------------ |
| `"wb"` | Write Binary |
| `"rb"` | Read Binary  |


### Write Content into Binary File (wb mode)

```python
# Writing into a binary file

data = "Hello Python Binary File"

with open("binary_file.bin", "wb") as file:
    file.write(data.encode())   # convert string to bytes
```

.encode() converts string into bytes (binary format)

### Read Content from Binary File (rb mode)

```python
# Reading from a binary file

with open("binary_file.bin", "rb") as file:
    content = file.read()
    print(content.decode())   # convert bytes back to string
```

### Read the content from the example-txt file and then writing the content into another file

```python
with open("example.txt","r") as file:
    content = file.read()
    print(content)
```

```python
# Read from example.txt and write into newfile.txt

with open("example.txt", "r") as file:
    content = file.read()   # read content

with open("newfile.txt", "w") as file:
    file.write(content)   # write content

print("Content copied successfully!")
```

### read total lines , total characters , total words from a file

```python
# Open the file in read mode
with open("example.txt", "r") as file:
    content = file.read()

# Total characters
total_characters = len(content)

# Total words
total_words = len(content.split())

# Total lines
total_lines = len(content.splitlines())

print("Total Lines:", total_lines)
print("Total Words:", total_words)
print("Total Characters:", total_characters)
```
