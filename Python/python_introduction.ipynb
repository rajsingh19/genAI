# What is python?


Python is a Type of programming language which we
can use for communicating with computers or
technology

Python is a high-level, interpreted programming language (Python runs line by line using an interpreter, so you don’t need to compile code before running it.)

## Syntax for python language

Syntax refers to the set of rules that defines the combination of symbols that are considered to be correctly
stuructured programs in a language

## Semantics of python language 

Semantics of python langugae refers to the meaning
or interpretation of symbols , characters and
commands in a langugae . It is about what the code i$
supposed to do when it runs .

## Python syntax rule

### 1️⃣ Indentation 

Python uses indentation (spaces) instead of {} to define blocks of code.

```python
if 5 > 2:
    print("Five is greater than two")
```

❌ Wrong (no indentation):

```python
#if 5 > 2:
#print("Five is greater than two")
```

### 2️⃣ No Semicolons Required

You don’t need ; at the end of statements.

```python
x = 10
y = 20
print(x + y)
```

### 3️⃣ Case Sensitivity

Python is case-sensitive.

```python
name = "John"
print(name)   # Works
print(Name)   # Error
```

### 4️⃣ Variables (No Type Declaration Needed)

You don’t declare variable types explicitly.

```python
age = 25          # integer
price = 19.99     # float
name = "Alice"    # string
```

### 5️⃣ Comments

Single-line comment:

```python
# This is a comment
```

Multi-line comment:

```python
"""
This is a
multi-line comment
"""
```

### 6️⃣ Proper Use of Colons (:)

A colon : is required after:

if

for

while

def

class

```python
for i in range(5):
    print(i)
```

### 7️⃣ Function Definition

```python
def greet(name):
    print("Hello", name)

greet("John")
```

### 9️⃣ String Quotes

You can use single or double quotes:

```python
text1 = "Hello"
text2 = 'World'
print(text1 + " " + text2)
```
