# Data Structure

A data structure is a way of organizing and storing data so it can be accessed and modified efficiently.

In Python, data structures help you:

Store multiple values

Organize data

Perform operations easily (search, sort, update, delete)

## üîπ Types of Data Structures in Python

Python mainly has two categories:

Built-in Data Structures

User-defined Data Structures

### üîπ 1Ô∏è‚É£ Built-in Data Structures

These are already available in Python.

#### ‚úÖ 1. List

Ordered

Mutable (can change)

Allows duplicate values

```python
numbers = [10, 20, 30, 40]

print(numbers[0])   # 10

numbers.append(50)  # Adding element
print(numbers)
```

#### ‚úÖ 2. Tuple

Ordered

Immutable (cannot change)

Allows duplicates

```python
coordinates = (5, 10)

print(coordinates[1])   # 10
```

#### ‚úÖ 3. Set

Unordered

Mutable

No duplicate values

```python
my_set = {1, 2, 3, 3}

print(my_set)   # {1, 2, 3}
```

#### ‚úÖ 4. Dictionary (dict)

Stores data in key-value pairs

Mutable

Keys must be unique

```python
student = {
    "name": "Alice",
    "age": 22
}

print(student["name"])   # Alice
```

### üîπ 2Ô∏è‚É£ User-Defined Data Structures

These are created by programmers.

#### ‚úÖ 1. Stack (LIFO ‚Äì Last In First Out)

Can be implemented using a list.

```python
stack = []

stack.append(10)
stack.append(20)
stack.pop()

print(stack)   # [10]
```

#### ‚úÖ 2. Queue (FIFO ‚Äì First In First Out)

```python
from collections import deque

queue = deque()

queue.append(10)
queue.append(20)
queue.popleft()

print(queue)   # deque([20])
```

#### ‚úÖ 3. Linked List (Custom Implementation)

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
```
