# EXCEPTION HANDLING

An exception is an error that happens during program execution.

Exception handling prevents the program from crashing.

## Example Without Exception Handling

```python
a = 10
b = 0

print(a / b)   # ‚ùå Error (ZeroDivisionError)
```

## Basic Exception Handling (try-except)

```python
try:
    a = 10
    b = 0
    print(a / b)
except:
    print("Error occurred!")
```

Program does not crash ‚úÖ

## Types of Exception Handling in Python

#### 1Ô∏è‚É£ try - except

```python
try:
    num = int(input("Enter number: "))
except:
    print("Invalid input!")
```

#### 2Ô∏è‚É£ try - except with specific error

```python
try:
    a = 10 / 0
except ZeroDivisionError:
    print("Cannot divide by zero")
```

#### 3Ô∏è‚É£ Multiple except blocks

```python
try:
    num = int(input("Enter number: "))
    result = 10 / num
except ValueError:
    print("Invalid number")
except ZeroDivisionError:
    print("Cannot divide by zero")
```

#### 4Ô∏è‚É£ try - except - else

üëâ else runs if no error occurs

```python
try:
    a = 10
    b = 2
    print(a / b)
except ZeroDivisionError:
    print("Error")
else:
    print("Division successful")
```

#### 5Ô∏è‚É£ try - except - finally

üëâ finally always runs (error or not)

```python
try:
    f = open("file.txt")
except FileNotFoundError:
    print("File not found")
finally:
    print("Execution finished")
```

#### 6Ô∏è‚É£ Raising Exception (raise)

```python
age = int(input("Enter age: "))

if age < 18:
    raise ValueError("Age must be 18 or above")
```

### Common Built-in Exceptions

| Exception         | Meaning                |
| ----------------- | ---------------------- |
| ZeroDivisionError | Divide by zero         |
| ValueError        | Wrong value type       |
| TypeError         | Wrong data type        |
| IndexError        | Invalid index          |
| KeyError          | Invalid dictionary key |
| FileNotFoundError | File not found         |
