# Data-type

Data types are classification of the data that tells the compiler or the interpreter that what
kind of values we can store into a variable and how much amount bf memory will be consumed to
store the data and the operations it can perform.

## üîπ 1Ô∏è‚É£ Numeric Data Types

#### ‚úÖ 1. int (Integer)

Whole numbers (positive or negative).

```python
x = 10
y = -5
print(type(x))
```

#### ‚úÖ 2. float (Decimal Numbers)

Numbers with decimal points.

```python
price = 99.99
print(type(price))
```

 #### ‚úÖ 3. complex (Complex Numbers)

Numbers with real and imaginary parts.

```python
z = 2 + 3j
print(type(z))
```

## üîπ 2Ô∏è‚É£ Text Type

#### ‚úÖ 4. str (String)

Used to store text.

```python
name = "Alice"
message = 'Hello World'
print(type(name))
```

## üîπ 3Ô∏è‚É£ Sequence Types

#### ‚úÖ 5. list

Ordered, changeable (mutable), allows duplicates.

```python
numbers = [1, 2, 3, 4]
print(numbers[0])   # Access first element
```

#### ‚úÖ 6. tuple

Ordered, but NOT changeable (immutable).

```python
coordinates = (10, 20)
print(coordinates[1])
```

#### ‚úÖ 7. range

Represents a sequence of numbers.

```python
r = range(5)
for i in r:
    print(i)
```

## üîπ 4Ô∏è‚É£ Set Types

#### ‚úÖ 8. set

Unordered, no duplicate values.

```python
my_set = {1, 2, 3, 3}
print(my_set)
```

#### ‚úÖ 9. frozenset

Same as set, but immutable.

```python
fs = frozenset({1, 2, 3})
print(fs)
```

## üîπ 5Ô∏è‚É£ Mapping Type

#### ‚úÖ 10. dict (Dictionary)

Stores data in key-value pairs.

```python
student = {
    "name": "John",
    "age": 20
}

print(student["name"])
```

## üîπ 6Ô∏è‚É£ Boolean Type

#### ‚úÖ 11. bool

Represents True or False.

```python
x = True
y = False
print(type(x))
```

## üîπ 7Ô∏è‚É£ None Type

#### ‚úÖ 12. NoneType

Represents no value.

```python
x = None
print(type(x))
```
