# Variables in Python


A variable is a name used to store data in a program.

In Python, you do not need to declare the type of a variable.

‚úÖ Example:

```python
name = "Alice"
age = 20
price = 99.99
```

Here:

name stores a string

age stores an integer

price stores a float

### üîπ Rules for Variable Names

‚úî Can contain letters, numbers, underscore _

‚úî Must start with a letter or underscore

‚ùå Cannot start with a number

‚úî Case-sensitive (age and Age are different)

```python
student_name = "John"   # valid
_age = 25               # valid
#2name = "Sam"           # ‚ùå invalid
```

### üîπ Changing Variable Value

```python
x = 10
x = 20   # value changed
print(x)
```

## print() Function

print() is used to display output on the screen.

### ‚úÖ Basic Example:

```python
print("Hello World")
```

### üîπ Printing Multiple Values

```python
name = "Anna"
age = 22
print("Name:", name, "Age:", age)
```

### üîπ Using f-strings 

```python
name = "Tom"
age = 25
print(f"My name is {name} and I am {age} years old.")
```

## input() Function

input() is used to take user input from the keyboard.

### ‚úÖ Basic Example:

```python
name = input("Enter your name: ")
print("Hello", name)
```

### üîπ Converting Input to Integer

To use numbers, convert using int():

```python
age = int(input("Enter your age: "))
print(age + 5)
```

### üéØ Combined Example (Variables + input + print)

```python
name = input("Enter your name: ")
age = int(input("Enter your age: "))

print(f"Hello {name}")
print(f"Next year you will be {age + 1} years old.")
```
