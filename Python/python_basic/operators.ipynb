# Operators

Operators are symbols used to perform operations on variables and values.

## üîπ 1Ô∏è‚É£ Arithmetic Operators

Used to perform mathematical operations.

| Operator | Meaning             | Example  |
| -------- | ------------------- | -------- |
| `+`      | Addition            | `5 + 3`  |
| `-`      | Subtraction         | `5 - 3`  |
| `*`      | Multiplication      | `5 * 3`  |
| `/`      | Division            | `5 / 2`  |
| `%`      | Modulus (remainder) | `5 % 2`  |
| `**`     | Exponent (power)    | `2 ** 3` |
| `//`     | Floor division      | `5 // 2` |


‚úÖ Example:

```python
a = 10
b = 3

print(a + b)   # 13
print(a - b)   # 7
print(a * b)   # 30
print(a / b)   # 3.3333
print(a % b)   # 1
print(a ** b)  # 1000
print(a // b)  # 3
```

## üîπ 2Ô∏è‚É£ Comparison Operators

Used to compare two values. Result is True or False.

| Operator | Meaning               |
| -------- | --------------------- |
| `==`     | Equal                 |
| `!=`     | Not equal             |
| `>`      | Greater than          |
| `<`      | Less than             |
| `>=`     | Greater than or equal |
| `<=`     | Less than or equal    |


‚úÖ Example:

```python
x = 5
y = 10

print(x == y)   # False
print(x != y)   # True
print(x > y)    # False
print(x < y)    # True
```

## üîπ 3Ô∏è‚É£ Assignment Operators

Used to assign values to variables.

| Operator | Example  | Same As     |
| -------- | -------- | ----------- |
| `=`      | `x = 5`  | assign      |
| `+=`     | `x += 3` | `x = x + 3` |
| `-=`     | `x -= 3` | `x = x - 3` |
| `*=`     | `x *= 2` | `x = x * 2` |
| `/=`     | `x /= 2` | `x = x / 2` |
| `%=`     | `x %= 2` | `x = x % 2` |


‚úÖ Example:

```python
x = 5
x += 3
print(x)   # 8
```

## üîπ 4Ô∏è‚É£ Logical Operators

Used to combine conditional statements.

| Operator | Meaning                                |
| -------- | -------------------------------------- |
| `and`    | True if both conditions are true       |
| `or`     | True if at least one condition is true |
| `not`    | Reverse the result                     |


‚úÖ Example:

```python
a = 10
b = 5

print(a > 5 and b > 2)   # True
print(a > 20 or b > 2)   # True
print(not(a > 5))        # False
```

## üîπ 5Ô∏è‚É£ Identity Operators

Used to compare memory locations.

| Operator | Meaning         |
| -------- | --------------- |
| `is`     | Same object     |
| `is not` | Not same object |


‚úÖ Example:

```python
x = [1, 2]
y = x
z = [1, 2]

print(x is y)      # True
print(x is z)      # False
```

## üîπ 6Ô∏è‚É£ Membership Operators

Used to test if a value exists in a sequence.

| Operator | Meaning              |
| -------- | -------------------- |
| `in`     | Value exists         |
| `not in` | Value does not exist |


‚úÖ Example:

```python
numbers = [1, 2, 3, 4]

print(2 in numbers)      # True
print(5 not in numbers)  # True
```

## üîπ 7Ô∏è‚É£ Bitwise Operators

Used to perform operations on binary numbers.

| Operator | Meaning     |    |
| -------- | ----------- | -- |
| `&`      | AND         |    |
| `        | `           | OR |
| `^`      | XOR         |    |
| `~`      | NOT         |    |
| `<<`     | Left shift  |    |
| `>>`     | Right shift |    |


‚úÖ Example:

```python
a = 5   # 101 in binary
b = 3   # 011 in binary

print(a & b)   # 1
print(a | b)   # 7
print(a ^ b)   # 6
```
